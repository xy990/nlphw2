```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# RNN Hidden Size
```

```python
RNN_hidden100 = pd.read_csv('RNN_hidden100.csv')
RNN_hidden200 = pd.read_csv('RNN_hidden200.csv')
RNN_hidden300 = pd.read_csv('RNN_hidden300.csv')
CNN_hidden100_kernel3 = pd.read_csv('CNN_hidden100_kernel3.csv')
CNN_hidden200_kernel3 = pd.read_csv('CNN_hidden200_kernel3.csv')

CNN_hidden300_kernel3 = pd.read_csv('CNN_hidden300_kernel3.csv')
CNN_hidden300_kernel2 = pd.read_csv('CNN_hidden300_kernel2.csv')
CNN_hidden300_kernel5 = pd.read_csv('CNN_hidden300_kernel5.csv')
CNN_hidden300_kernel7 = pd.read_csv('CNN_hidden300_kernel7.csv')


RNN_hidden300_drop0 = pd.read_csv('RNN_hidden300_drop0.csv')
RNN_hidden300_drop1 = pd.read_csv('RNN_hidden300_drop1.csv')
RNN_hidden300_drop5 = pd.read_csv('RNN_hidden300_drop5.csv')

CNN_hidden300_kernel3_drop1 = pd.read_csv('CNN_hidden300_kernel3_drop1.csv')
CNN_hidden300_kernel3_drop5 = pd.read_csv('CNN_hidden300_kernel3_drop5.csv')

CNN_hidden300_kernel3_drop8 = pd.read_csv('CNN_hidden300_kernel3_drop8.csv')

```

```python
rnn_train_acc_hidden100 = RNN_hidden100['train_accs']
rnn_val_acc_hidden100 = RNN_hidden100['val_accs']
rnn_train_acc_hidden200 = RNN_hidden200['train_accs']
```

```python
x= range(len(RNN_hidden100))
y1 = RNN_hidden100['train_accs']
y2 = RNN_hidden200['train_accs']
y3 = RNN_hidden300['train_accs']

plt.plot(x, y1,linestyle='dashed',marker='o',color='b')
plt.plot(x, y2,linestyle='dashed',marker='o',color='r')
plt.plot(x, y3,linestyle='dashed',marker='o',color='g')

plt.legend(['hidden_size= 100', 'hidden_size= 200','hidden_size= 300'], bbox_to_anchor=(1, 1.05))
plt.xlabel('number of epoch steps')
plt.ylabel("train accuracy")
plt.title(' Train Accuracy of Hidden Size')
plt.savefig("Train_Acc_hiddensize_rnn.jpeg")
plt.show()
```

```python

x= range(len(RNN_hidden100))
y1 = RNN_hidden100['val_accs']
y2 = RNN_hidden200['val_accs']
y3 = RNN_hidden300['val_accs']
plt.plot(x, y1,linestyle='dashed',marker='o',color='b')
plt.plot(x, y2,linestyle='dashed',marker='o',color='r')
plt.plot(x, y3,linestyle='dashed',marker='o',color='g')

plt.legend(['hidden_size= 100', 'hidden_size= 200','hidden_size= 300'],loc='upper left', bbox_to_anchor=(1, 1.05))
plt.xlabel('number of epoch steps')
plt.ylabel("validation accuracy")
plt.title('Validation Accuracy of Hidden Size')
plt.savefig("Val_Acc_hiddensize_rnn.jpeg")
plt.show()
```
